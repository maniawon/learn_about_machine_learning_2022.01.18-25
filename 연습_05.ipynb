# Nowhere → <font color='red'>Now here

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 읽어오기
path = 'https://bit.ly/AdmissionFile'
data = pd.read_csv(path)

# 확인
data.head()
```

```python
# x, y 분리
target = 'ADMIT'
x = data.drop(target, axis=1)
y = data[target] # == data.loc[:, target]
```

```python
# 학습용, 평가 데이터 분리
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,
                                                   random_state=2022)
```

```python
# 불러오기
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# 선언하기
model = DecisionTreeClassifier(max_depth=3) # max_depth=None : 디폴트 값

# 검증하기
cv_score = cross_val_score(model, x_train, y_train, cv=10) # cv=5 : 디폴트 값

# 확인
print(cv_score)
print('평균:', cv_score.mean())
```

```python
# 선언하기
depth = []
score = []
for d in range(1, 11) :
    model = DecisionTreeClassifier(max_depth=d)
    cv_score = cross_val_score(model, x_train, y_train, cv=10)
    score = score + cv_score.tolist()
    depth = depth + [d] * 10

# 데이터프레임 만들기
result = pd.DataFrame()
result['depth'] = depth
result['score'] = score

# 확인
result.sort_values(by='score', ascending=False)
```

> 리스트 + 리스트 = 리스트합치기

```python
tmp = result.groupby(by='depth', as_index=False)['score'].mean()
tmp.sort_values(by='score', ascending=False)
```
